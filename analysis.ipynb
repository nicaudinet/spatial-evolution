```python
histories_mm = opt_lattice_mm.take(0)['histories']
present_strategies_mm = opt_lattice_mm.take(0)['present_strategies']

quasi_extict_limit = 100

all_last_strats_mm = []

for hist, present_strats in zip(histories_mm, present_strategies_mm):
    last_gen = hist[:,-1]
    last_strats = present_strats[last_gen > quasi_extict_limit]
    all_last_strats_mm.append(last_strats)
```

```python
flat_strats_mm = []

for trial in all_last_strats_mm:
    for x in trial:
        flat_strats_mm.append(str(x))
label, count = np.unique(flat_strats_mm, return_counts=True)
plt.bar(label, count)
```

## all vs all with mutations and mistakes

```python
filenames = [
    'data/try_intermediate_0_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_1_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_2_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_3_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_4_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_5_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_6_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_7_1673608700_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_0_1673613600_all_mut_0.01_mist_0.01_rounds_100.npy',
    'data/try_intermediate_0_1673614488_all_mut_0.01_mist_0.01_rounds_100.npy',
]

histories_mm = []
present_strategies_mm = []

for f in filenames:
    np_file = np.load(f, allow_pickle=True)
    histories_mm.append(np_file.take(0)['history'])
    present_strategies_mm.append(np.array(np_file.take(0)['present_strategies']))

quasi_extict_limit = 100

all_last_strats_mm = []

for hist, present_strats in zip(histories_mm, present_strategies_mm):
    last_gen = hist[:,-1]
    last_strats = present_strats[last_gen > quasi_extict_limit]
    all_last_strats_mm.append(last_strats)

flat_strats_mm = []

for trial in all_last_strats_mm:
    for x in trial:
        flat_strats_mm.append(str(x))

label, count = np.unique(flat_strats_mm, return_counts=True)

plt.bar(label, count)
```
